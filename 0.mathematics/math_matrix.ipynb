### 1. 矩阵与向量乘法
$\vec y= A\vec x$
* 若$x$ 为$n$ 维向量，$A_{m\times n}$， $y$ 为$m$维向量，则可以看做空间变化
* 若$A_{n \times n}$，$x$ 与 $y$ 均为$n$维的话，可以视为旋转或者平移变化，$A$ 为对应的变换矩阵

### 2. 矩阵的秩
* 矩阵$A_{m\times n}$中，任取$k$行$k$列，不改变这$k^2$个元素，在$A$中的次序,得到$k$阶方阵，称为矩阵$A$的$k$阶子式。
    * 显然,矩阵$A_{m\times n}$的k阶子式有 $C_m^kC_n^k$ 个。
* 设在矩阵$A$中有一个不等于0的$r$阶子式$D$，且所有$r+1$阶子式(如果存在的话)全等于0，那么，$D$称为矩阵$A$的最高阶非零子式，$r$称为矩阵$A$的秩，记做$R(A)=r$
    * ${n\times n}$的可逆矩阵,秩为$n$
    * 可逆矩阵又称满秩矩阵
    * 矩阵的秩等于它行(列)向量组的秩
* 对于$n$元线性方程组$Ax=b$
    * 无解的充要条件是$R(A)<R(A,b)$
    * 有唯一解的充要条件是$R(A)=R(A,b)=n$
    * 有无限多解的充要条件是$R(A)=R(A,b)<n$

### 3. 正交矩阵

* 若$n$阶矩阵$A$满足$A^TA =I$，成$A$为正交矩阵,简称正交阵。
* $A$是正交阵的充要条件: $A$的列(行)向量都是单位向量，且两两正交。
    * $A$是正交阵，$x$ 为向量，则 $Ax$称作正交变换。
    * 正交变换不改变向量长度

### 4. 特征值与特征向量
* $A$是$n$阶矩阵，若数$λ$和$n$维非$0$列向量$x$满足 $Ax=\lambda x$，那么，数$\lambda$称为$A$的特征值,$x$称为$A$的对应于特征值$\lambda$的特征向量。
    * 根据定义，立刻得到$(A-\lambda I)x = 0$，令关于$\lambda $ 的多项式$|A-\lambda I|$为0，方程$|A-\lambda I|= 0$的根为$A$的特征值。将根$\lambda_0$带入方程组$(A-\lambda I)x = 0$，求得到的非零解，即$\lambda_0$ 对应的特征向量。
* 设$n$阶矩阵$A=(a_{ij})$ 的特征值为$\lambda_1, \lambda_2,\dots ,\lambda_n$，则
    * $\lambda_1 + \lambda_2 + \dots + \lambda_n = a_{11} + a_{22} + \dots + a_{nn}$
    * $\lambda_1  \lambda_2  \dots  \lambda_n =|A|$，矩阵$A$主行列式的元素和，称作矩阵$A$的迹
* 设$\lambda_1,  \lambda_2,  \dots  \lambda_m$是方阵$A$的$m$个特征值，$p_1, p_2,\dots,p_m$ 是依次与之对应的特征向量，若$\lambda_1,  \lambda_2,  \dots  \lambda_m$各不相等，则$p_1, p_2,\dots,p_m$ 线性无关
* 实对称阵的特征值是实数
* 实对称阵不同特征值的特征向量正交
* 设$A$为$n$阶对称阵，则必有正交阵$P$，使得
    $$P^{-1}AP = P^TAP = \Lambda $$
    * $\Lambda$是以$A$的$n$个特征值为对角元的对角阵
    * 该变换称为"合同变换"，$A$和$\Lambda$互为合同矩阵


### 5. 正定矩阵
* 对于$n$阶方阵$A$，若任意$n$阶向量$x$，都有$x^TAx\gt0$，则称$A$是正定阵。
    * 若条件变成$x^TAx\ge 0$，则$A$称作半正定阵
    * 类似还有负定阵，半负定阵。
* 给定任意$m\times n$的矩阵$A$, 可证明$A^TA$一定是半正定方阵。
* 正定阵的判定
    * 对称阵$A$为正定阵
    * $A$的特征值都为正
    * $A$的顺序主子式大于0
    * 以上三个命题等价

### 6. 矩阵分解
* $QR$分解
    * 对于$m\times n$的列满秩矩阵$A$, 必有: $A_{m \times n} = Q_{m \times n} \cdot R_{n \times n}$
    * 其中, $Q^TQ =I$ (即列正交矩阵)，$R$为非奇异上三角矩阵。当要求$R$的对角线元素为正时，该分解唯一。
    * 该分解为$QR$分解。可用于求解矩阵$A$的特征值、$A$的逆等问题。
* $LFM$ (Latent Factor Model)
    * 对于$K$个隐变量,得: $A_{m \times n} = U_{m \times k} \cdot V_{n \times k}^T$，可以应用于推荐系统
        * $U_{m \times k}$ 可以视为用户对于商品的偏好，$U_{n\times k}$ 可以视为商品的特征矩阵。
        * 可以用于计算用户对于具体商品的偏好程度
* $SVD$奇异值分解
$$A_{m \times n}=U_{m \times m}\Sigma_{m \times n} V^T_{n \times n}$$
 * 由于奇异值由大到小排列，所以取前$k$ 个可以用来做主成分分析

* $QR$分解计算$n$阶方阵$A$的特征值:   
    $$A=QR \Rightarrow A_1 = Q^TAQ = RQ$$
    $$\dots$$
    $$A_k =Q_kR_K \Rightarrow A_{k+1} = R_kQ_k$$
    $$\dots$$
    $$A_K \rightarrow diag \{\lambda_1, \lambda_2, \dots, \lambda_n\}$$

### 7.向量导数
* 矩阵$A_{m\times n}$, 列向量 $x_{n\times 1}$，则$\vec y = A \vec x$，其中$y$为$m×1$的列向量，则向量偏导公式如下：
$$\frac{\partial A \vec x }{\partial \vec x} = A^T$$
$$\frac{\partial A \vec x }{\partial \vec x^T} = A$$
$$\frac{\partial(\vec x A)}{\partial \vec x} = A$$
* $A$为$n\times n$的矩阵, $x$为$n\times 1$的列向量, 记 $\vec y = \vec x^T A \vec x$
$$\frac{\partial \vec x^T A \vec x }{\partial \vec x} = (A^T + A)\vec x$$
 * 若$A$为对阵矩阵，则
$$\frac{\partial \vec x^T A \vec x }{\partial \vec x} = 2A\vec x$$
